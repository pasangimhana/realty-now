# Data Cleanup for Employment Data

```python
grouped_data = employment_by_block.groupby(['Year', 'Area'])['Total Jobs'].sum().reset_index()
difference_data = []
years = grouped_data['Year'].unique()

for year in years:
    data_for_year = grouped_data[grouped_data['Year'] == year]
    city_of_melbourne_total = data_for_year[data_for_year['Area'] == 'City of Melbourne (total)']['Total Jobs'].values[0]
    rest_of_melbourne_total = data_for_year[data_for_year['Area'] != 'City of Melbourne (total)']['Total Jobs'].sum()
    difference = city_of_melbourne_total - rest_of_melbourne_total
    difference_data.append({
        'Year': year,
        'City of Melbourne (total)': city_of_melbourne_total,
        'Sum of Other Areas': rest_of_melbourne_total,
        'Difference': difference
    })
difference_df = pd.DataFrame(difference_data)
print(difference_df)
```

Looks like the City of Melbournce (total) dataset of the Area column is roughly the sum of Other Areas. So we can remove it.

```python
employment_by_block = employment_by_block[employment_by_block['Area'] != 'City of Melbourne (total)']
employment_by_block
```

# Clustering for Areas and Employment to Get Residential Details

```python
residential_dwellings
```

```python
employment_by_area
```

```python
employment_dwelling = pd.merge(residential_dwellings.groupby(['Year', 'Area'])['Number of Dwellings'].sum().reset_index(),
                       employment_by_area,
                       on=['Year', 'Area'],
                       how='outer')

employment_dwelling.fillna(0, inplace=True)

employment_dwelling
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np

# Preprocessing
features = employment_dwelling.drop(columns=['Year', 'Area'])
features.fillna(0, inplace=True)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Clustering
optimal_n_clusters = 4
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42)
employment_dwelling['Cluster'] = kmeans.fit_predict(features_scaled)

# PCA for visualization
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

# Plotting
plt.figure(figsize=(10, 6))
scatter = plt.scatter(features_pca[:, 0], features_pca[:, 1], c=employment_dwelling['Cluster'], cmap='viridis', alpha=0.6)
plt.title('K-Means Clustering of Employment and Dwellings')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')

# Selectively label some points (change the criteria as needed)
unique_clusters = np.unique(employment_dwelling['Cluster'])
for cluster in unique_clusters:
    # Get a few representative points from each cluster
    cluster_points = employment_dwelling[employment_dwelling['Cluster'] == cluster].sample(min(5, len(employment_dwelling[employment_dwelling['Cluster'] == cluster])))
    for i, txt in cluster_points.iterrows():
        plt.annotate(txt['Area'],
                     (features_pca[i, 0], features_pca[i, 1]),
                     fontsize=8,
                     alpha=0.7,
                     bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

plt.show()
```

```python
centroids = kmeans.cluster_centers_
centroids_df = pd.DataFrame(scaler.inverse_transform(centroids), columns=features.columns)

print('Centroids:')
print(centroids_df)

cluster_counts = employment_dwelling['Cluster'].value_counts()
print('Cluster sizes:')
print(cluster_counts)

cluster_means = employment_dwelling['Cluster'].mean()
print('Mean feature values per cluster:')
print(cluster_means)

area_cluster_distribution = employment_dwelling.groupby(['Area', 'Cluster']).size().reset_index(name='Counts')
print('Area and Cluster distribution:')
print(area_cluster_distribution)

year_cluster_distribution = employment_dwelling.groupby(['Year', 'Cluster']).size().reset_index(name='Counts')
print('Year and Cluster distribution:')
print(year_cluster_distribution)
```

# Time Series Analysis of the Data upto 2016

```python
print(house_price_and_jobs.info())
print(house_price_and_jobs.describe())
print(house_price_and_jobs.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

house_price_and_jobs['Area'] = house_price_and_jobs['Area'].astype('category')
house_price_and_jobs['Type'] = house_price_and_jobs['Type'].astype('category')

correlation_matrix = house_price_and_jobs.select_dtypes(include=['number']).corr()

plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=house_price_and_jobs, x='Year', y='Median Price', estimator='mean')
plt.title('Median Price Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Median Price')
plt.grid()
plt.show()
```

```python
area_price = house_price_and_jobs.groupby('Area')['Median Price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=area_price.sort_values('Median Price', ascending=False), x='Median Price', y='Area')
plt.title('Mean Median Price by Area')
plt.xlabel('Mean Median Price')
plt.ylabel('Area')
plt.grid()
plt.show()
```

```python
annual_prices = house_price_and_jobs.groupby('Year')['Median Price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=annual_prices, x='Year', y='Median Price')
plt.title('Annual Median Prices')
plt.xlabel('Year')
plt.ylabel('Median Price')
plt.grid()
plt.show()
```
